```
from pyspark.sql import SparkSession
```

```
spark = SparkSession.builder.appName('TechNights Veri Analizi').getOrCreate()
```

```
rawDS = spark.read.csv('/FileStore/tables/orderstechnights.csv', header="true", inferSchema="true")
```

```
rawDS.show()
```

```
rawDS.printSchema()
```

```
citygroupDS = rawDS.groupBy("CITY").count()
```

```
descCityCountDS = citygroupDS.orderBy("count", ascending=False).limit(5)
```

```
descCityCountDS.show()
```

```
display(descCityCountDS)
```

```
from pyspark.sql.functions import *
```

```
wdateDS = rawDS.withColumn("to_date",date_format("DATE_","yyyy-MM-dd"))
```

```
itemCountDS = wdateDS.groupBy("to_date","ITEMNAME").count()
```

```
itemCountDescDS = itemCountDS.orderBy("count", ascending=False).limit(100)
```

```
display(itemCountDescDS)
```

```
totalpriceByUser = rawDS.groupBy("USERID").sum("ITEM_UNITPRICE").withColumnRenamed("sum(ITEM_UNITPRICE)","totalprice")
```

```
resultPriceByUserDS = totalpriceByUser.orderBy("totalprice", ascending=False).limit(5)
```

```
resultPriceByUserDS.withColumn("total",bround("totalprice",2)).drop("totalprice").show()
```

```
rawDS.createOrReplaceTempView("orders")
```
